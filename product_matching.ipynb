```python
# Doing a left join on flipkart and amazon datasets using uniq_id as primary key. Left join will take all values from Flipkart and match with Amazon
data = pd.merge(flipkart,amazon,on='uniq_id', how='left')
```

```python
# droppimg unnecessary columns

data=data.drop(["crawl_timestamp_x","product_url_x","is_FK_Advantage_product_x","is_FK_Advantage_product_y","product_category_tree_x","pid_x","image_x","description_x","product_rating_x",
           "overall_rating_x","product_category_tree_y","pid_y","image_y","description_y","product_rating_y",
           "overall_rating_y","brand_y","product_specifications_y","brand_x",
           "product_specifications_x","crawl_timestamp_y","product_url_y","uniq_id"],axis=1)


```
